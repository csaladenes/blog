```
import cv2
import numpy as np
```

```
def vid_resize(from_path,to_path,pix):
  from_path=from_path.replace('"','')
  to_path=to_path.replace('"','')
  cap = cv2.VideoCapture(from_path)
  fps = cap.get(cv2.CAP_PROP_FPS)
  first=True
  success, frame = cap.read()
  #this section is to experiment with resolutions that don't break the tensor shapes
  # height , width , layers =  frame.shape
  # new_h=int(height*a)-1 #change thsi between 0 and -3 until it works
  # new_w=int(width*a)
  # if new_h%2==1: new_h+=1
  # if new_w%2==1: new_w+=1
  new_w=pix[0]
  new_h=pix[1]
  print(new_w,'x',new_h)
  fourcc = cv2.VideoWriter_fourcc(*'XVID')
  out = cv2.VideoWriter(to_path,fourcc, fps, (new_w,new_h))
      
  while success:
      b = cv2.resize(frame,(new_w,new_h),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
      out.write(b)
      success, frame = cap.read()
      
  cap.release()
  out.release()
  cv2.destroyAllWindows()
  return
```
